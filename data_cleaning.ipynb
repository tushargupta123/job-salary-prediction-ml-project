## what we can do for cleaning
1. remove unnamed column
2. remove data with salary estimates as -1 , take average of salary estimates
3. job description seems take skills form it
4. make yes or no field for checking if headquarters and location are same 
5. average out revenue
6. take only count of competitors
7. remove \n after company name
8. take upper limiy of size 
9. take only ratings greater than 0 rest put to 0
10. instead of founded get age of company

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

```python
li = []
for index,row in df.iterrows():
    start = row['Salary Estimate'].split('(')[0].split('$')[1].split('K')[0]
    end = row['Salary Estimate'].split('(')[0].split('$')[2].split('K')[0]
    li.append((int(''.join(char for char in start if char.isdigit())) + int(''.join(char for char in end if char.isdigit())))/2)
df['Salary Estimate'] = li
```

```python
li = []
for index, row in df.iterrows():
    revenue = row['Revenue'].split('$')
    if len(revenue) >= 3:
        start = revenue[1].split()[0]
        end = revenue[2].split()[0]
        li.append((int(''.join(char for char in start if char.isdigit())) + int(''.join(char for char in end if char.isdigit())))/2)
    else:
        li.append(None)

df['Revenue'] = li
```

```python
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
```

```python
df['Size'].value_counts()
```

```python
df = df[df['Size'] != '-1']
```

```python
li = []
for index,row in df.iterrows():
    if row['Size']!='Unknown':
        size = row['Size'].split('to')[0].split('employees')[0]
        li.append((int(''.join(char for char in size if char.isdigit()))))
    else:
        li.append(None)
df['Size'] = li
```

```python
li = []
for index,row in df.iterrows():
    if row['Competitors']!='-1':
        li.append(len(row['Competitors'].split(',')))
    else:
        li.append(0)
df['Competitors'] = li
```

```python
li = []
for index,row in df.iterrows():
    if row['Rating']<0:
        li.append(0)
    else:
        li.append(row['Rating'])
df['Rating'] = li
```

```python
df['Founded'] = 2023 - df['Founded']
```

```python
df.rename(columns={'Founded':'Age'},inplace=True)
```

```python
df['Is_Headquarter'] = (df['Headquarters']==df['Location'])
```

```python
df['Python'] = df['Job Description'].apply(lambda x:1 if 'python' in x.lower() else 0)
df['Spark'] = df['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
df['Aws'] = df['Job Description'].apply(lambda x:1 if 'aws' in x.lower() else 0)
df['Excel'] = df['Job Description'].apply(lambda x:1 if 'excel' in x.lower() else 0)
```
